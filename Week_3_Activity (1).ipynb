### 4.Read the given data set and answer the following questions. Show the steps taken to produce your final answer.

1. Display the Glucose level and the Age for the samples having the index ranging from 45 to 60.
2. How many patients in the data are having the blood pressure level above 120?
3. Find the minimum and maximum value of insulin level in the data.
4. Create a histogram of "Age".
5. Create a scatter plot of "BMI" vs "Glucose".

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('diabetes_model.csv')
```

```python
data = pd.DataFrame(data)
```

```python
data
```

```python
# 1.Display the Glucose level and the Age for the samples having the index ranging from 45 to 60.
```

```python
data[["Glucose","Age"]][45:61]
```

```python
#2.How many patients in the data are having the blood pressure level above 120?
```

```python
bp_lvl = (data['BloodPressure'] > 120)
```

```python
bp_lvl.sum()
```

```python
#3.Find the minimum and maximum value of insulin level in the data.
```

```python
print("Minimum value of insulin =",data['Insulin'].min())
print("Maximum value of insulin =",data['Insulin'].max())
```

```python
#4.Create a histogram of "Age".
```

```python
plt.figure(figsize=(8,6))
plt.hist(data['Age'],rwidth = 0.6,color = 'Red')
plt.title('Age of Diabetes patients',fontsize = 15)
plt.xticks()
plt.yticks()
plt.xlabel('Age of Patients',fontsize = 10)
plt.ylabel('Number of patients',fontsize = 10)
```

```python
#Create a scatter plot of "BMI" vs "Glucose".
```

```python
plt.figure(figsize = (8,8))
plt.scatter(data['BMI'],data['Glucose'],s = 15,c = 'violet')
plt.title('Scatter plot of BMI vs Glucose',fontsize = 16)
plt.xlabel('BMI',fontsize = 15)
plt.ylabel('Glucose',fontsize = 15)
```
